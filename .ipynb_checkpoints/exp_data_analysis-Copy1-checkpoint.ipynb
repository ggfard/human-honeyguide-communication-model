```python
def plot_waveform_spectrogram(y, sr, title, segments=None, desired_fmax=None, filename_prefix="waveform_spectrogram"):
    """Plots the waveform and mel spectrogram of an audio signal with optional segment markers."""
    fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

    # Waveform
    librosa.display.waveshow(y, sr=sr, ax=axs[0], color='royalblue', label='Waveform')
    axs[0].set_title('Waveform')
    axs[0].set_ylabel('Amplitude')

    # Add segment onset lines if provided
    if segments:
        for start_time, end_time in segments:
            axs[0].axvline(x=start_time, color='r', linestyle='--', linewidth=0.9, label='Segment Onset')
            axs[0].axvline(x=end_time, color='g', linestyle='--', linewidth=0.9, label='Segment End')
            axs[1].axvline(x=start_time, color='white', linestyle='--', linewidth=0.9) # Spectrogram onset

    # To avoid duplicate labels if you add them in a loop (legend for first occurrence only)
    handles, labels = axs[0].get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    if by_label:
        axs[0].legend(by_label.values(), by_label.keys(), loc='upper right')

    # Mel Spectrogram
    plot_fmax = desired_fmax if desired_fmax is not None else sr / 2.0
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=plot_fmax)
    S_DB = librosa.amplitude_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_DB, sr=sr, x_axis='time', y_axis='mel', ax=axs[1], fmax=plot_fmax)
    axs[1].set_title('Mel Spectrogram')
    axs[1].set_ylabel('Frequency (Mel)')
    axs[1].set_xlabel('Time (s)')

    fig.colorbar(img, ax=axs[1], format='%+2.0f dB')
    plt.suptitle(title)
    plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust rect to make space for suptitle
    plt.savefig(os.path.join(OUTPUT_FOLDER, f"{filename_prefix}.png"))
```

```python
def analyze_call_audio(audio_path, species_type, 
                       top_db_threshold=DEFAULT_TOP_DB, 
                       min_call_dur=MIN_CALL_DURATION_DEFAULT, 
                       max_call_dur=MAX_CALL_DURATION_DEFAULT,
                       pre_post_roll_s=PRE_POST_ROLL,
                       pitch_fmin=None, pitch_fmax=None,
                       load_max_duration=None,
                       desired_spec_fmax=None,
                       plot_filename_suffix=""):
    
    """Loads, analyzes, and extracts features from an audio file."""
    
    print(f"Processing {species_type} audio file: {audio_path}")
    try:
        y, sr = librosa.load(audio_path, sr=None, duration=load_max_duration)
        actual_loaded_duration = librosa.get_duration(y=y, sr=sr)
        print(f"Loaded audio: Actual Duration={actual_loaded_duration:.2f}s (Max requested: {load_max_duration if load_max_duration else 'Full'}s), SR={sr}Hz")

        print(f"Detecting call segments (top_db={top_db_threshold}, min_dur={min_call_dur}s, max_dur={max_call_dur}s)...")
        call_segments_times = detect_call_segments(y, sr, 
                                                   top_db=top_db_threshold, 
                                                   min_duration=min_call_dur,
                                                   max_duration=max_call_dur)
        print(f"Found {len(call_segments_times)} potential call segments.")

        plot_waveform_spectrogram(y, sr, 
                                  title=f"{species_type} Audio: {os.path.basename(audio_path)}", 
                                  segments=call_segments_times,
                                  desired_fmax=desired_spec_fmax,
                                  filename_prefix=f"{species_type.lower()}_waveform_spectrogram{plot_filename_suffix}")
        
        if not call_segments_times:
            print(f"No calls detected for {species_type}. Skipping feature extraction.")
            return []

        all_features = []
        print("Extracting features for each segment...")
        
        if pitch_fmin is None:
            pitch_fmin = HUMAN_FMIN_DEFAULT if species_type.lower() == "human" else BIRD_FMIN_DEFAULT
        if pitch_fmax is None:
            pitch_fmax = HUMAN_FMAX_DEFAULT if species_type.lower() == "human" else BIRD_FMAX_DEFAULT

        for i, (start_t, end_t) in enumerate(call_segments_times):
            start_sample_padded = max(0, librosa.time_to_samples(start_t, sr=sr) - int(pre_post_roll_s * sr))
            end_sample_padded = min(len(y), librosa.time_to_samples(end_t, sr=sr) + int(pre_post_roll_s * sr))
            y_segment = y[start_sample_padded:end_sample_padded]
            
            features = extract_features_for_segment(y_segment, sr, pitch_fmin, pitch_fmax)
            features["start_time_s"] = start_t
            features["end_time_s"] = end_t
            features["original_duration_s"] = end_t - start_t
            features["audio_path"] = audio_path # Add audio_path to features
            features["total_audio_duration_loaded"] = actual_loaded_duration # Add loaded duration
            
            all_features.append(features)
            pitch_display = f"{features['mean_pitch_hz']:.2f}Hz" if not np.isnan(features['mean_pitch_hz']) else "nanHz"
            print(f"  Segment {i+1} ({start_t:.2f}s - {end_t:.2f}s): "
                  f"OrigDur={features['original_duration_s']:.2f}s, "
                  f"Pitch={pitch_display}, "
                  f"RMS={features['rms_energy']:.3f}")
        
        return all_features

    except Exception as e:
        print(f"Error processing file {audio_path}: {e}")
        return []
```

```python
def print_mean_features(features_list, species_name):
    
    """Calculates and prints mean statistics for a list of feature dictionaries."""
    
    num_segments = len(features_list)
    if num_segments > 0:
        mean_orig_duration = np.mean([f['original_duration_s'] for f in features_list])
        
        pitches = [f['mean_pitch_hz'] for f in features_list if 'mean_pitch_hz' in f and not np.isnan(f['mean_pitch_hz'])]
        mean_pitch = np.mean(pitches) if pitches else float('nan')
        num_voiced_segments = len(pitches)

        mean_max_amplitude = np.mean([f['max_amplitude'] for f in features_list])
        mean_rms_energy = np.mean([f['rms_energy'] for f in features_list])

        print(f"\nMean Statistics for {species_name} Calls ({num_segments} segments):")
        print(f"  Mean Original Duration: {mean_orig_duration:.3f}s")
        pitch_stat_display = f"{mean_pitch:.2f}Hz (from {num_voiced_segments} voiced segments)" if not np.isnan(mean_pitch) else f"N/A (0 voiced segments of {num_segments})"
        print(f"  Mean Pitch: {pitch_stat_display}")
        print(f"  Mean Max Amplitude: {mean_max_amplitude:.3f}")
        print(f"  Mean RMS Energy: {mean_rms_energy:.3f}")
    else:
        print(f"\nNo segments found for {species_name} to calculate mean statistics.")
```

```python
def plot_duration_histogram(features_list, species_name, bins=20, color='mediumseagreen', filename_prefix="duration_histogram"):
    """Plots a histogram of call segment durations."""
    if not features_list:
        print(f"No segments found for {species_name} to plot duration histogram.")
        return

    durations = [feat['original_duration_s'] for feat in features_list if 'original_duration_s' in feat]
    
    if not durations:
        print(f"No durations found in features for {species_name}.")
        return
        
    plt.figure(figsize=(6, 4))
    plt.hist(durations, bins=bins, color=color, edgecolor='black')
    plt.title(f"Distribution of {species_name} Call Segment Durations")
    plt.xlabel("Duration (s)")
    plt.ylabel("Frequency (Number of Segments)")
    plt.grid(axis='y', alpha=0.75)
    plt.savefig(os.path.join(OUTPUT_FOLDER, f"{filename_prefix}.png"))
```

```python
def detect_call_segments(y, sr, top_db=DEFAULT_TOP_DB, 
                       min_duration=MIN_CALL_DURATION_DEFAULT, 
                       max_duration=MAX_CALL_DURATION_DEFAULT):
    """Detects non-silent segments (potential calls) in an audio signal."""
    intervals_samples = librosa.effects.split(y, top_db=top_db, frame_length=2048, hop_length=512)
    
    call_segments = []
    for start_sample, end_sample in intervals_samples:
        start_time = librosa.samples_to_time(start_sample, sr=sr)
        end_time = librosa.samples_to_time(end_sample, sr=sr)
        duration = end_time - start_time
        
        if min_duration <= duration <= max_duration:
            call_segments.append((start_time, end_time))
            
    return call_segments
```

```python
def extract_features_for_segment(y_segment, sr, fmin, fmax):
    """Extracts features for a single audio segment."""
    duration = len(y_segment) / sr
    max_amplitude = np.max(np.abs(y_segment)) if len(y_segment) > 0 else 0.0
    rms_energy = np.sqrt(np.mean(y_segment**2)) if len(y_segment) > 0 else 0.0
    
    mean_pitch_hz = np.nan
    std_pitch_hz = np.nan # Added std_pitch_hz
    if len(y_segment) > 0:
        f0, voiced_flag, _ = librosa.pyin(y_segment, fmin=fmin, fmax=fmax, sr=sr, fill_na=np.nan)
        if np.any(voiced_flag):
            voiced_f0 = f0[voiced_flag]
            mean_pitch_hz = np.nanmean(voiced_f0)
            std_pitch_hz = np.nanstd(voiced_f0) # Calculate std_f0

    return {
        "duration_s": duration,
        "max_amplitude": max_amplitude,
        "rms_energy": rms_energy,
        "mean_pitch_hz": mean_pitch_hz,
        "std_f0": std_pitch_hz # Return std_f0
    }
```

```python
def plot_event_timeline(human_calls, bird_calls, total_duration_s, title, filename="event_timeline.png"):
    """Plots human and bird calls on a timeline."""
    fig, ax = plt.subplots(figsize=(15, 4))

    y_positions = {'human': 1.0, 'bird': 0.5}
    colors = {'human': 'royalblue', 'bird': 'forestgreen'}
    
    all_calls = []
    if human_calls:
        for call in human_calls:
            all_calls.append({'start': call['start_time_s'], 
                              'end': call['end_time_s'], 
                              'type': 'human'})\
                              
    if bird_calls:
        for call in bird_calls:
            all_calls.append({'start': call['start_time_s'], 
                              'end': call['end_time_s'], 
                              'type': 'bird'})

    if not all_calls:
        print("No calls to plot on timeline.")
        ax.set_xlim(0, total_duration_s if total_duration_s > 0 else 1)
        ax.set_ylim(0, 1.5)
        ax.set_yticks([y_positions['human'], y_positions['bird']])
        ax.set_yticklabels(['Human', 'Bird'])
        ax.set_xlabel("Time (s)")
        plt.title(title)
        plt.grid(axis='x', linestyle=':')
        plt.savefig(os.path.join(OUTPUT_FOLDER, filename))
        #plt.close(fig)
        return

    for call_event in all_calls:
        start = call_event['start']
        end = call_event['end']
        call_type = call_event['type']
        duration = end - start
        
        rect = patches.Rectangle((start, y_positions[call_type] - 0.2), duration, 0.4, 
                                 linewidth=1, edgecolor=colors[call_type], 
                                 facecolor=colors[call_type], alpha=0.7)
        ax.add_patch(rect)

    ax.set_ylim(0, 1.5)
    ax.set_xlim(0, total_duration_s)
    ax.set_yticks([y_positions['human'], y_positions['bird']])
    ax.set_yticklabels(['Human', 'Bird'])
    ax.set_xlabel("Time (s)")
    plt.title(title)
    plt.grid(True, linestyle=':')
    
    human_patch = patches.Patch(color=colors['human'], label='Human Call', alpha=0.7)
    bird_patch = patches.Patch(color=colors['bird'], label='Bird Call', alpha=0.7)
    ax.legend(handles=[human_patch, bird_patch], loc='upper right')
    plt.savefig(os.path.join(OUTPUT_FOLDER, filename))
    #plt.close(fig)
```

```python
def calculate_response_latencies(source_calls, responder_calls, max_latency_s=5.0):
    """
    Calculates response latencies from the end of a source call
    to the start of the next responder call, within a max_latency_s window.
    """
    latencies = []
    if not source_calls or not responder_calls:
        return latencies

    sorted_responder_calls = sorted(responder_calls, key=lambda x: x['start_time_s'])
    
    responder_idx = 0
    for src_call in source_calls:
        src_end_time = src_call['end_time_s']
        
        while responder_idx < len(sorted_responder_calls) and \
              sorted_responder_calls[responder_idx]['start_time_s'] < src_end_time:
            responder_idx += 1
        
        temp_idx = responder_idx
        while temp_idx < len(sorted_responder_calls):
            resp_call = sorted_responder_calls[temp_idx]
            latency = resp_call['start_time_s'] - src_end_time
            if 0 <= latency <= max_latency_s:
                latencies.append(latency)
                break
            elif latency > max_latency_s:
                break
            temp_idx += 1
            
    return latencies
```

```python
def plot_latency_histogram(latencies, title, bins=20, filename_prefix="latency_histogram"):
    """Plots a histogram of response latencies."""
    if not latencies:
        print(f"No latencies to plot for: {title}.")
        return
        
    plt.figure(figsize=(8, 5))
    plt.hist(latencies, bins=bins, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel("Latency (s)")
    plt.ylabel("Frequency")
    plt.grid(axis='y', alpha=0.75)
    plt.savefig(os.path.join(OUTPUT_FOLDER, f"{filename_prefix}.png"))
```

```python
def calculate_conditional_calling(source_events, target_events, response_window_s=1.0):
    """
    Calculates the probability of a target event occurring within response_window_s
    AFTER a source event has ended.
    P(Target Call | Source Call ended in previous response_window_s)
    """
    if not source_events:
        return 0.0 # Return 0 probability if no source events to avoid None/errors
    num_source_events_followed_by_target = 0
    for src_event in source_events:
        src_end_time = src_event['end_time_s']
        response_window_end = src_end_time + response_window_s
        found_target_in_window = False
        for tgt_event in target_events:
            if src_end_time < tgt_event['start_time_s'] <= response_window_end:
                found_target_in_window = True
                break
        if found_target_in_window:
            num_source_events_followed_by_target += 1
            
    return num_source_events_followed_by_target / len(source_events)
```

```python

response_window_for_cond_prob = 1.0 # e.g., probability of response within 1 second after other's call ends

prob_bird_after_human = calculate_conditional_calling(all_human_whistle_features, all_bird_features, response_window_for_cond_prob)
prob_human_after_bird = calculate_conditional_calling(all_bird_features, all_human_whistle_features, response_window_for_cond_prob)

print(f"\nConditional Calling Probabilities (within {response_window_for_cond_prob}s window):")
print(f"  P(Bird Call | Human Call Ended): {prob_bird_after_human:.2f}")
print(f"  P(Human Call | Bird Call Ended): {prob_human_after_bird:.2f}")

# --- 2.4 Moving Window Leadership Analysis ---
analysis_window_size = 5.0  # How much data to consider for each leadership score calculation
analysis_step_size = 0.5    # How much to slide the window
cond_resp_window = 1.0
```

```python
def moving_window_leadership_analysis(human_calls, bird_calls, total_duration_s, 
                                     window_size_s=10.0, step_size_s=2.0, 
                                     conditional_response_window_s=1.0, filename="leadership_dynamics.png"):
    """
    Calculates a leadership score in moving windows.
    Leadership Score = P(Bird|Human) - P(Human|Bird)
    """
    window_start_times = np.arange(0, total_duration_s - window_size_s + step_size_s, step_size_s)
    leadership_scores = []
    window_mid_points = []

    if not window_start_times.size:
        print("Interaction too short for moving window analysis with these parameters.")
        return [], []

    for t_start in window_start_times:
        t_end = t_start + window_size_s
        window_mid_points.append(t_start + window_size_s / 2)

        h_calls_in_window = [c for c in human_calls if t_start <= c['start_time_s'] < t_end]
        b_calls_in_window = [c for c in bird_calls if t_start <= c['start_time_s'] < t_end]

        if not h_calls_in_window and not b_calls_in_window:
            leadership_scores.append(0)
            continue
        
        prob_bird_after_human = calculate_conditional_calling(
            h_calls_in_window, b_calls_in_window, conditional_response_window_s
        )
        prob_human_after_bird = calculate_conditional_calling(
            b_calls_in_window, h_calls_in_window, conditional_response_window_s
        )
        
        score = prob_bird_after_human - prob_human_after_bird
        leadership_scores.append(score)
        
    if window_mid_points: # Check if there are any points to plot
        plt.figure(figsize=(8, 6))
        plt.plot(window_mid_points, leadership_scores, marker='o', linestyle='-', color='purple')
        plt.axhline(0, color='grey', linestyle='--', linewidth=0.8)
        plt.xlabel("Time (center of analysis window, s)")
        plt.ylabel("Leadership Score (P(B|H) - P(H|B))")
        plt.title(f"Vocal Leadership Dynamics\\n(Window: {analysis_window_size}s, Step: {analysis_step_size}s, RespWin: {cond_resp_window}s)")
        plt.ylim(-1.1, 1.1)
        plt.grid(True, linestyle=':')
        plt.savefig(os.path.join(OUTPUT_FOLDER, filename))
        
    return window_mid_points, leadership_scores
```

```python
window_times, leadership_dynamic = moving_window_leadership_analysis(
    all_human_whistle_features, all_bird_features, total_interaction_duration,
    window_size_s=analysis_window_size,
    step_size_s=analysis_step_size,
    conditional_response_window_s=cond_resp_window,
    filename="vocal_leadership_dynamics.png"
)
if window_times:
    print(f"\nVocal Leadership Dynamics: {leadership_dynamic} (Plotted as 'vocal_leadership_dynamics.png')")
    print("Positive score: Bird potentially leading vocally. Negative score: Human potentially leading.")
```

# Section 3: Model Parameter Informing

```python
# 3.1. Informing delta_w (Difference in Intrinsic Frequencies)
# This assumes Hazda whistles are the primary human call for phase sync in this context.
mean_human_whistle_f0 = np.nanmean([f['mean_pitch_hz'] for f in current_hazda_whistle_features if not np.isnan(f['mean_pitch_hz'])])
mean_bird_f0 = np.nanmean([f['mean_pitch_hz'] for f in current_bird_features if not np.isnan(f['mean_pitch_hz'])])

if not np.isnan(mean_human_whistle_f0) and not np.isnan(mean_bird_f0):
    delta_w_estimate_hz = np.abs(mean_human_whistle_f0 - mean_bird_f0)
    # Convert to rad/s for consistency if your model expects rad/s. omega = 2 * pi * f
    delta_w_estimate_rad_s = delta_w_estimate_hz * 2 * np.pi
    print(f"\nEstimated delta_w (from Hazda whistles and Bird F0): {delta_w_estimate_rad_s:.4f} rad/s")
    print("  How it informs the model: Directly constrains the `delta_w` parameter in `dphi_dt = delta_w - 2 * K[t] * np.sin(phi[t])`. A larger `delta_w` implies a greater intrinsic frequency mismatch, making synchrony harder to achieve and maintain, potentially requiring higher `K_max` or more robust `beta`.")
    
    # Normalization example (requires defining a Max_real_delta_w)
    # For demonstration, let's assume Max_real_delta_w from all possible vocalizations could be 1000 Hz or ~6283 rad/s
    MAX_REAL_DELTA_W_RAD_S = 6283.0 # Example: assuming 1000 Hz difference is the max
    normalized_delta_w = delta_w_estimate_rad_s / MAX_REAL_DELTA_W_RAD_S
    print(f"  Normalized delta_w (assuming max real-world delta_w = {MAX_REAL_DELTA_W_RAD_S:.0f} rad/s): {normalized_delta_w:.4f} (for model range [0,1])")
else:
    print("\nCould not estimate delta_w due to missing mean F0 for human whistles or bird calls.")

# 3.2. Informing noise_strength_base (Baseline Noise)
# Use standard deviation of F0 as a proxy for vocal variability/noise.
# Take an average of whistle and bird F0 std. dev, as grunts might be less relevant for "oscillator" noise.
human_whistle_f0_std = np.nanmean([f['std_f0'] for f in current_hazda_whistle_features if not np.isnan(f['std_f0'])]) if current_hazda_whistle_features else np.nan
bird_f0_std = np.nanmean([f['std_f0'] for f in current_bird_features if not np.isnan(f['std_f0'])]) if current_bird_features else np.nan

if not np.isnan(human_whistle_f0_std) and not np.isnan(bird_f0_std):
    noise_strength_proxy = (human_whistle_f0_std + bird_f0_std) / 2
    print(f"\nProxy for intrinsic noise strength (from F0 std dev): {noise_strength_proxy:.4f} Hz")
    print("  How it informs the model: The `noise_strength_base` parameter directly influences the stochastic term in the `dphi_dt` equation. Higher baseline noise makes synchrony less stable and harder to initiate. This can represent inherent variability in call production or background environmental noise. You might need to calibrate its scale by running simulations and comparing phase stability.")
else:
    print("\nCould not estimate noise_strength_proxy due to missing F0 standard deviations.")

# 3.3. Informing alpha (Modulation Strength on x dynamics)
print("\nInforming alpha (Modulation Strength on x dynamics):")
print("  This is more qualitative but can be quantified if you have behavioral data linking call types to cooperation.")
# Example qualitative insight from previous analysis:
if current_human_grunt_features and current_hazda_whistle_features:
    mean_grunt_f0_std = np.nanmean([f['std_f0'] for f in current_human_grunt_features if not np.isnan(f['std_f0'])])
    if not np.isnan(human_whistle_f0_std) and not np.isnan(mean_grunt_f0_std):
        if human_whistle_f0_std < mean_grunt_f0_std:
            print("  Hazda whistles appear more tonally stable than grunts (lower F0 std. dev.). If tonal signals are known to facilitate better synchrony and cooperative progression in honeyguides, this could suggest a higher `alpha` for whistle-based interactions compared to grunt-based ones. This parameter scales the boost to `x` growth from synchrony (`1 + alpha * cos(phi)`).")

# Quantitative proxy if available from interaction data:
if prob_bird_after_human > prob_human_after_bird: # Bird responds more reliably to human than vice versa
    print("  Observed: Birds respond more reliably to human calls than humans respond to bird calls (P(B|H) > P(H|B)). This might imply `alpha` (human participation boost by synchrony) is relatively high, as bird responses are crucial for human task progress. The model could explore if higher `alpha` values lead to more stable human participation when synchrony is achieved.")
else:
    print("  Observed: Humans respond more reliably to bird calls than birds respond to human calls (P(H|B) > P(B|H)). This might imply `alpha` (human participation boost by synchrony) is relatively lower, as human participation might not be as strongly reinforced by bird calls, or other factors are more dominant.")

# 3.4. Informing 'r' (Growth rate for x) - Human Participation
# Proxy for human call rate/participation
if all_human_whistle_features:
    total_human_calls = len(all_human_whistle_features)
    total_human_audio_duration_for_rate = all_human_whistle_features[0]['total_audio_duration_loaded'] if all_human_whistle_features else 0
    human_call_rate = total_human_calls / total_human_audio_duration_for_rate if total_human_audio_duration_for_rate > 0 else 0
    print(f"\nProxy for human call rate/participation: {human_call_rate:.4f} calls/s")
    print("  How it informs the model: A higher rate of calls might suggest higher intrinsic `r` (growth rate for human participation `x`). If humans are more persistent in calling, `x` might grow faster for a given `S_acc`. This parameter directly influences how quickly the task `x` progresses.")
else:
    print("\nNo human call features to estimate human call rate.")

# 3.5. Informing 'beta' (Growth rate for K) and 'delta' (Decay rate for K) - Coupling Strength
print("\nInforming 'beta' (Growth rate for K) and 'delta' (Decay rate for K) - Coupling Strength:")
print("  These are harder to get directly from isolated calls but can be indirectly informed.")
# Use mean call durations and ICIs for qualitative inference
mean_human_call_duration = np.nanmean([f['original_duration_s'] for f in all_human_whistle_features]) if all_human_whistle_features else np.nan
mean_bird_call_duration = np.nanmean([f['original_duration_s'] for f in all_bird_features]) if all_bird_features else np.nan

mean_human_to_bird_ici = np.mean(human_to_bird_latencies) if human_to_bird_latencies else np.nan
mean_bird_to_human_ici = np.mean(bird_to_human_latencies) if bird_to_human_latencies else np.nan

if not np.isnan(mean_human_call_duration) and not np.isnan(mean_bird_call_duration):
    print(f"  Mean Human Call Duration: {mean_human_call_duration:.3f}s")
    print(f"  Mean Bird Call Duration: {mean_bird_call_duration:.3f}s")
    print(f"  Mean H->B Inter-Call Latency (Response): {mean_human_to_bird_ici:.3f}s")
    print(f"  Mean B->H Inter-Call Latency (Response): {mean_bird_to_human_ici:.3f}s")
    print("  How they inform the model: Shorter response latencies and ICIs might suggest that coupling needs to be reinforced more frequently, or decays faster if not. This could imply a higher `delta`. Conversely, very sustained, regular calls could support a higher `beta` (faster coupling build-up). These are qualitative inferences, and the absolute values would need fitting.")
else:
    print("  Call duration/ICI data insufficient for inferences on beta/delta.")

# 3.6. Informing 'thresh_cos' (Threshold for S_acc)
print("\nInforming 'thresh_cos' (Threshold for S_acc):")
if not np.isnan(human_whistle_f0_std) and not np.isnan(bird_f0_std):
    mean_f0_std_combined = (human_whistle_f0_std + bird_f0_std) / 2
    print(f"  Combined Mean F0 Std Dev (Human Whistle & Bird): {mean_f0_std_combined:.4f} Hz")
    print("  How it informs the model: Lower `std_f0` suggests more consistent, less noisy vocalizations. If the communication is expected to be very precise to be effective, `thresh_cos` (which determines how tightly phases must align for `S_acc` accumulation) could be set higher. A higher `thresh_cos` means more stringent synchrony is required to progress the task.")
else:
    print("  F0 standard deviation data insufficient for inferences on thresh_cos.")
```
